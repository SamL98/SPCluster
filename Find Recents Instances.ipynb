```python
import pandas as pd
```

```python
recents = pd.read_csv('/users/samlerner/documents/spotify/recently_played.csv')

strip_quotes = lambda x: x[1:-1] if (x[0] == '\'' and x[-1] == '\'') else x
recents['artist'] = recents['artist'].apply(strip_quotes)
recents['title'] = recents['title'].apply(strip_quotes)
recents['timestamp'] = recents['timestamp'].apply(lambda t: int(t)).astype(int)

recents = recents.sort_values('timestamp')
```

```python
times = recents.timestamp
durations = recents.duration / 1000
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(times, [1]*len(times), width=durations)
plt.show()
```
